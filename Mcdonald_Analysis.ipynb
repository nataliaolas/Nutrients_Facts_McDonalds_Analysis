1. Ranking McDonalds items with the most calories

```python
meal = df.query('Category != "Beverages" and Category !=  "Coffee & Tea" and Category != "Smoothies & Shakes"') #generate all meals without drinks 
meal.Category.unique()
meals_cal = pd.DataFrame({'Item': meal.Item, 'Calories': meal.Calories}) # Select only Items and Calories columns
meals_sort = meals_cal.sort_values('Calories', ascending=True) # Sort by calories
meals_sort.plot.barh(x='Item', y='Calories', figsize= (10,40), color='red') 
plt.show()
```

2. How many calories does the average McDonald's meal contain?

```python
brkfst = df.loc[df.Category =='Breakfast']
bandp = df.loc[df.Category == 'Beef & Pork']
chandfish = df.loc[df.Category == 'Chicken & Fish']
sal = df.loc[df.Category == 'Salads']
snands = df.loc[df.Category == 'Snacks & Sides']
dess = df.loc[df.Category == 'Desserts']
```

```python
average_calories_category = round(brkfst['Calories'].mean(axis=0),1),round(bandp['Calories'].mean(axis=0),1), round(chandfish['Calories'].mean(axis=0),1), round(sal['Calories'].mean(axis=0),1), round(snands['Calories'].mean(axis=0),1),round(dess['Calories'].mean(axis=0),1) 
index = ['Breakfast','Beef & Pork', 'Chicken & Fish', 'Salads','Snacks & Sides', 'Desserts']
avg_cat= pd.DataFrame({'Avg Calories': average_calories_category}, index=index)
fig = avg_cat.plot.bar(rot=0, color='red', figsize =(10,8), title='Average calories for each category')
fig.set_xlabel('Categories')
fig.set_ylabel('Calorie')

for bar in fig.patches:
    fig.annotate(str(bar.get_height()), (bar.get_x() * 1.045, bar.get_height() * 1.012))
```

4. Plots for nutritions, compared with maximum amount recommended for an adult diet
plt.figure(figsize=(16,3))

```python
#Average amount of saturated fat compared to the maximum amount recommended for an adult diet
plt.figure(figsize=(16,3))
sns.barplot(x=df.groupby('Category')['Saturated Fat'].mean().index, y=df.groupby('Category')['Saturated Fat'].mean().values)
plt.title('Average amount of saturated fat compared to the maximum amount recommended for an adult diet ')
plt.ylabel('Saturate Fat')
plt.xlabel('Category')
plt.axhline(y=16,color='red',linewidth=4) #16 g for one adult person per day
```

```python
#Average amount of sugares compared to the maximum amount recommended for an adult diet
plt.figure(figsize=(16,3))
sns.barplot(x=df.groupby('Category')['Sugars'].mean().index, y=df.groupby('Category')['Sugars'].mean().values)
plt.title('Average amount of sugars compared to the maximum amount recommended for an adult diet')
plt.ylabel('Sugar')
plt.xlabel('Category')
plt.axhline(y=50,color='red',linewidth=4) #50 g for one adult person per day
```

```python
#Average amount of cholesterol compared to the maximum amount recommended for an adult diet
plt.figure(figsize=(16,3))
sns.barplot(x=df.groupby('Category')['Cholesterol'].mean().index, y=df.groupby('Category')['Cholesterol'].mean().values)
plt.title('Average amount of cholesterols compared to the maximum amount recommended for an adult diet')
plt.ylabel('Cholesterol')
plt.xlabel('Category')
plt.axhline(y=180,color='red',linewidth=4) #180 mg for one adult person per day
```

5. Correlation between nutritions

```python
nutrition = [
 'Calories',
 'Calories from Fat',
 'Total Fat',
 'Saturated Fat',
 'Trans Fat',
 'Cholesterol',
 'Sodium',
 'Carbohydrates',
 'Dietary Fiber',
 'Sugars',
 'Protein',
 'Vitamin A (% Daily Value)',
 'Vitamin C (% Daily Value)',
 'Calcium (% Daily Value)',
 'Iron (% Daily Value)'
 ]

correl = df[nutrition].corr()
fig = plt.figure(figsize = (15,15))
sns.heatmap(correl, square=True, cmap='vlag', annot=True)
```
